```python
from cogent3.app import io, evo

loader = io.load_aligned(format="fasta", moltype="dna")
aln = loader("data/primate_brca1.fasta")
```

```python
model = evo.model("GNC", tree="data/primate_brca1.tree")
result = model(aln)
result
```
